# Create Population

```python
def createChromosome(finalList, finalD):
    chromosome = []
    for _ in range(35):  #35 rows in the timetable = number of rooms
        gene = []
        for _ in range(6):  #6 columns in the timetable
            gene.append(random.randint(0, len(finalList) - 1))
        chromosome.append(gene)
    
    for row in chromosome:
        for key in range(len(row)):
            row[key] = list(finalD[row[key]])
    return chromosome

#create population with diversity
def createPopulation(numOfParents, finalList, finalD):
    return [createChromosome(finalList, finalD) for _ in range(numOfParents)]
```

# Fitness Function

```python
def fitness(chromosome):
    punish = 0
    instructorDailyClasses = {}
    studentDailyClasses = {}

    for col in range(6):
        instructors = set()
        sections = set()
        for row in range(35):
            ins = chromosome[row][col][2]
            sec = chromosome[row][col][1]
            if ins in instructors:
                punish += 10  #penalty
            if sec in sections:
                punish += 10  #penalty
            instructors.add(ins)
            sections.add(sec)
            
            #track daily classes
            if ins not in instructorDailyClasses:
                instructorDailyClasses[ins] = [0] * 6
            if sec not in studentDailyClasses:
                studentDailyClasses[sec] = [0] * 6
            instructorDailyClasses[ins][col] += 1
            studentDailyClasses[sec][col] += 1

    #penalize if any instructor or student has more than 4 classes in a day
    for dailyClasses in instructorDailyClasses.values():
        if max(dailyClasses) > 4:
            punish += 10  #penalty
    
    for dailyClasses in studentDailyClasses.values():
        if max(dailyClasses) > 4:
            punish += 10  #penalty
    
    score = 1 / (punish + 1)
    return score
```

# Crossover & Mutation

```python
def swapColumns(p1, p2, flag):
    if flag == 0:
        col = random.randint(0, 5)
        for i in range(35):
            p1[i][col], p2[i][col] = p2[i][col], p1[i][col]
    elif flag == 1:
        cols = random.sample(range(6), random.randint(1, 3))
        for col in cols:
            for i in range(35):
                p1[i][col], p2[i][col] = p2[i][col], p1[i][col]
    elif flag == 2:
        for i in range(3):
            for row in range(35):
                p1[row][i], p2[row][i + 3] = p2[row][i + 3], p1[row][i]
    return p1, p2

def swapRows(p1, p2):
    if random.random() < 0.5:
        row = random.randint(0, 34)
        p1[row], p2[row] = p2[row], p1[row]
    else:
        rows = random.sample(range(35), random.randint(1, 30))
        for row in rows:
            p1[row], p2[row] = p2[row], p1[row]
    return p1, p2

def crossover(parents):
    p1, p2 = parents
    p1, p2 = swapColumns(p1, p2, random.randint(0, 2))
    child1, child2 = swapRows(p1, p2)
    return child1, child2

def mutation(individual, probability, savedPop):
    if random.random() < probability:
        toMutateWith = random.choice(savedPop)
        individual, _ = swapColumns(individual, toMutateWith, random.randint(0, 2))
    if random.random() < probability:
        toMutateWith = random.choice(savedPop)
        individual, _ = swapRows(individual, toMutateWith)
    return individual
```

# Genetic Algorithm

```python
def selectionTournament(population):
    parents = []
    for _ in range(2):
        tournament = [random.choice(population) for _ in range(5)]
        parents.append(max(tournament, key=fitness))
    return parents

def getMaxFitnessValue(population):
    return max(population, key=fitness)

def geneticAlgorithm(finalList, finalD, generations=500, populationSize=50):
    population = createPopulation(populationSize, finalList, finalD)
    savedPop = population.copy()
    maxFitnessDict = {}

    for generation in range(generations):
        newPopulation = []
        for _ in range(populationSize // 2):
            parents = selectionTournament(population)
            child1, child2 = crossover(parents)
            child1 = mutation(child1, 0.1, savedPop)
            child2 = mutation(child2, 0.1, savedPop)
            newPopulation.extend([child1, child2])
        
        bestIndividual = getMaxFitnessValue(population)
        newPopulation[1] = bestIndividual
        population = newPopulation
        maxFitnessDict[fitness(bestIndividual)] = bestIndividual
        sortedDict = dict(sorted(maxFitnessDict.items(), reverse=True))
        firstKey = list(sortedDict.keys())[0]
        population[0] = maxFitnessDict[firstKey]

        print(f"Generation {generation + 1}, Max Fitness: {firstKey}, Best Individual's Fitness: {fitness(bestIndividual)}")

        if fitness(bestIndividual) == 1:
            print("Solution found!")
            return bestIndividual
    
    return getMaxFitnessValue(population)

bestTimetable = geneticAlgorithm(finalList, finalD)
```
